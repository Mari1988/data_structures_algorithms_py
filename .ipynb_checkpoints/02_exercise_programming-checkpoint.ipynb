```python
def factors(n):
    
    f_list = []
    for i in range(1,n+1):
        if n % i == 0:
            f_list = f_list + [i]
    return(f_list)

def isprime(z):
    
    f_list_z = factors(z)
    if f_list_z == [1,z]:
        return(z)
    else:
        return(0)

def primepartition(m):
    
    if m > 0:
        p_list = []
        for i in range(2,m):
            if isprime(i) != 0:
                p_list = p_list + [i]
        
        for ind1 in range(0,len(p_list)):
            e1 = p_list[ind1]
            if ind1 <= len(p_list)-1:
                for ind2 in range(ind1+1,len(p_list)):
                    if e1+p_list[ind2] == m:
                        return(True)
        return(False)
    else:
        return(False)
    
def nestingdepth(s):
    count=0
    max_depth = 0
    for i in s:
        if i=="(":
            count += 1;
        elif i == ")" and count !=0:
            count -= 1;
        elif i == ")" and count == 0:
            return(-1)
        max_depth = max(max_depth,count)
        
    if count == 0:
        return(max_depth)
    else:
        return(-1)
    
def rotatelist(l,k):
    sub_l = l[:]
    for i in [0]*k:
        f_elem = sub_l[i]
        sub_l.pop(i)
        sub_l = sub_l + [f_elem]
    return(sub_l)
```
